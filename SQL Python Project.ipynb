```python
# replace old column names
df.columns = df.columns.str.lower()
```

```python
df.head()
```

```python
# add underscores where necessary
# bad method: df.rename(columns = {'order id':'order_id', 'order date':'order_date'})
# a better method is:
df.columns.str.replace(' ','_') # this replaces all spaces with an underscore
```

```python
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
# calculated discount amount
df['discount'] = df['list_price']*df['discount_percent']/100
```

```python
# calculate sale price
df['sale_price'] = df['list_price'] - df['discount']
```

```python
df.head()
```

```python
# calculate profit
df['profit'] = df['sale_price'] - df['cost_price']
df.head()
```

```python
# drop columns no longer needed
df = df.drop(['cost_price','list_price','discount_percent'],axis = 1)
```

```python
# convert order date to datetime
df.dtypes
```

```python
df['order_date'] = pd.to_datetime(df['order_date'],format = '%Y-%m-%d')
df.dtypes
```
